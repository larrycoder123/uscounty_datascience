```python
import pandas as pd
```

```python
education_df = pd.read_csv("./datatest/education_county_2023.csv", dtype={"FIPS Code": str})

# Create new column "Pop25+!HighSchool+!Percentage"
# = HighSchool or Higher / Population 25+ years
education_df["Pop25+!HighSchool+!Percentage"] = education_df["Population 25 years and over!Cumulative!High school graduate or higher"] / education_df["Population 25 years and over"]

# Create new column "Pop25+!BachelorsDegree+!Percentage"
# = Bachelors Degree or Higher / Population 25+ years
education_df["Pop25+!BachelorsDegree+!Percentage"] = education_df["Population 25 years and over!Cumulative!Bachelors degree or higher"] / education_df["Population 25 years and over"]

# Drop columns that are no longer needed
education_df = education_df.drop(["Population 25 years and over!Cumulative!Bachelors degree or higher", "Population 25 years and over!Cumulative!High school graduate or higher", "Population 25 years and over"], axis=1)

education_df.head()
```

```python
internetaccess_df = pd.read_csv("./datatest/internetaccess_county_2023.csv", dtype={"FIPS Code": str})

# Create new column "Households!WithInternetSubscription!Percentage"
# = Households With an Internet Subscription / Total Households
internetaccess_df["Households!WithInternetSubscription!Percentage"] = internetaccess_df["Households!With an Internet Subscription"] / internetaccess_df["Total Households"]

# Drop columns that are no longer needed
internetaccess_df = internetaccess_df.drop(["Households!With an Internet Subscription", "Total Households"], axis=1)

internetaccess_df.head()
```

```python
meanincome_df = pd.read_csv("./datatest/meanincome_county_2023.csv", dtype={"FIPS Code": str})

# Rename column
meanincome_df.rename(columns={"Households!Mean Income": "Housholds!MeanIncome"}, inplace = True)

# Drop columns that are no longer needed
meanincome_df = meanincome_df.drop(["Total Households"], axis=1)

meanincome_df.head()
```

```python
fertilityrate_df = pd.read_csv("./datatest/fertilityrate_county_2023.csv", dtype={"FIPS Code": str})

# The general fertility rate (GFR) measures births per 1,000 women of childbearing age (15-50)
# Create new column "GeneralFertilityRate"
# = (Number of women with births in the past year / Total number of women aged 15-50)×1000
fertilityrate_df["GeneralFertilityRate"] = (fertilityrate_df["Women 15 to 50 years!Women with births in the last 12 months"] / fertilityrate_df["Total Women 15 to 50 years"]) * 1000


# Drop columns that are no longer needed
fertilityrate_df = fertilityrate_df.drop(["Women 15 to 50 years!Women with births in the last 12 months", "Total Women 15 to 50 years"], axis=1)

fertilityrate_df.head()
```

```python
householdsize_df = pd.read_csv("./datatest/householdsize_county_2023.csv", dtype={"FIPS Code": str})

# Rename column
householdsize_df.rename(columns={"AvgHouseholdSize": "Average Household Size"}, inplace = True)

householdsize_df.rename(columns={"Total Households": "Households!Total"}, inplace = True)

householdsize_df.head()
```

```python
industrycomposition_df = pd.read_csv("./datatest/industrycomposition_county_2023.csv", dtype={"FIPS Code": str})


# Calculate Percentages

industrycomposition_df["16+Employed!AgricultureForestryFishingHuntingMining!Percentage"] = industrycomposition_df["16+ Employed!Agriculture, forestry, fishing and hunting, and mining"] / industrycomposition_df["Total 16+ Employed"]

industrycomposition_df["16+Employed!Construction!Percentage"] = industrycomposition_df["16+ Employed!Construction"] / industrycomposition_df["Total 16+ Employed"]

industrycomposition_df["16+Employed!Manufacturing!Percentage"] = industrycomposition_df["16+ Employed!Manufacturing"] / industrycomposition_df["Total 16+ Employed"]

industrycomposition_df["16+Employed!Wholesaletrade!Percentage"] = industrycomposition_df["16+ Employed!Wholesale Trade"] / industrycomposition_df["Total 16+ Employed"]

industrycomposition_df["16+Employed!Retailtrade!Percentage"] = industrycomposition_df["16+ Employed!Retail Trade"] / industrycomposition_df["Total 16+ Employed"]

industrycomposition_df["16+Employed!TransportationWarehousingUtilities!Percentage"] = industrycomposition_df["16+ Employed!Transportation and warehousing, and utilities"] / industrycomposition_df["Total 16+ Employed"]

industrycomposition_df["16+Employed!Information!Percentage"] = industrycomposition_df["16+ Employed!Information"] / industrycomposition_df["Total 16+ Employed"]

industrycomposition_df["16+Employed!FinanceInsuranceRealestate!Percentage"] = industrycomposition_df["16+ Employed!Finance and insurance, and real estate and rental and leasing"] / industrycomposition_df["Total 16+ Employed"]

industrycomposition_df["16+Employed!ProfessionalScientificManagementAdministrativeServices!Percentage"] = industrycomposition_df["16+ Employed!Professional, scientific, and management, and administrative and waste management services"] / industrycomposition_df["Total 16+ Employed"]

industrycomposition_df["16+Employed!EducationHealthcareSocialassistance!Percentage"] = industrycomposition_df["16+ Employed!Educational services, and health care and social assistance"] / industrycomposition_df["Total 16+ Employed"]

industrycomposition_df["16+Employed!ArtsEntertainmentRecreationAccommodationFood!Percentage"] = industrycomposition_df["16+ Employed!Arts, entertainment, and recreation, and accommodation and food services"] / industrycomposition_df["Total 16+ Employed"]

industrycomposition_df["16+Employed!Otherservices!Percentage"] = industrycomposition_df["16+ Employed!Other services, except public administration"] / industrycomposition_df["Total 16+ Employed"]

industrycomposition_df["16+Employed!Publicadministration!Percentage"] = industrycomposition_df["16+ Employed!Public administration"] / industrycomposition_df["Total 16+ Employed"]

# Rename

industrycomposition_df.rename(columns={"Total 16+ Employed": "16+Employed!Total"}, inplace = True)

# Drop Columns 

industrycomposition_df = industrycomposition_df.drop(
    ["16+ Employed!Agriculture, forestry, fishing and hunting, and mining",
     "16+ Employed!Construction",
     "16+ Employed!Manufacturing",
     "16+ Employed!Wholesale Trade",
     "16+ Employed!Retail Trade",
     "16+ Employed!Transportation and warehousing, and utilities",
     "16+ Employed!Information",
     "16+ Employed!Finance and insurance, and real estate and rental and leasing",
     "16+ Employed!Professional, scientific, and management, and administrative and waste management services",
     "16+ Employed!Educational services, and health care and social assistance",
     "16+ Employed!Arts, entertainment, and recreation, and accommodation and food services",
     "16+ Employed!Other services, except public administration",
     "16+ Employed!Public administration"
     ],axis=1)
```
