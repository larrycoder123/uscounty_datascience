```python
import pandas as pd
```

```python
education_df = pd.read_csv("./datatest/education_county_2023.csv", dtype={"FIPS Code": str})

# Create new column "Pop25+!HighSchool+!Percentage"
# = HighSchool or Higher / Population 25+ years
education_df["Pop25+!HighSchool+!Percentage"] = education_df["Population 25 years and over!Cumulative!High school graduate or higher"] / education_df["Population 25 years and over"]

# Create new column "Pop25+!BachelorsDegree+!Percentage"
# = Bachelors Degree or Higher / Population 25+ years
education_df["Pop25+!BachelorsDegree+!Percentage"] = education_df["Population 25 years and over!Cumulative!Bachelors degree or higher"] / education_df["Population 25 years and over"]

# Drop columns that are no longer needed
education_df = education_df.drop(["Population 25 years and over!Cumulative!Bachelors degree or higher", "Population 25 years and over!Cumulative!High school graduate or higher", "Population 25 years and over"], axis=1)

education_df.head()
```

```python
internetaccess_df = pd.read_csv("./datatest/internetaccess_county_2023.csv", dtype={"FIPS Code": str})

# Create new column "Households!WithInternetSubscription!Percentage"
# = Households With an Internet Subscription / Total Households
internetaccess_df["Households!WithInternetSubscription!Percentage"] = internetaccess_df["Households!With an Internet Subscription"] / internetaccess_df["Total Households"]

# Drop columns that are no longer needed
internetaccess_df = internetaccess_df.drop(["Households!With an Internet Subscription", "Total Households"], axis=1)

internetaccess_df.head()
```

```python
meanincome_df = pd.read_csv("./datatest/meanincome_county_2023.csv", dtype={"FIPS Code": str})

# Rename column
meanincome_df.rename(columns={"Households!Mean Income": "Housholds!MeanIncome"}, inplace = True)

# Drop columns that are no longer needed
meanincome_df = meanincome_df.drop(["Total Households"], axis=1)

meanincome_df.head()
```

```python
fertilityrate_df = pd.read_csv("./datatest/fertilityrate_county_2023.csv", dtype={"FIPS Code": str})

# The general fertility rate (GFR) measures births per 1,000 women of childbearing age (15-50)
# Create new column "GeneralFertilityRate"
# = (Number of women with births in the past year / Total number of women aged 15-50)×1000
fertilityrate_df["GeneralFertilityRate"] = (fertilityrate_df["Women 15 to 50 years!Women with births in the last 12 months"] / fertilityrate_df["Total Women 15 to 50 years"]) * 1000


# Drop columns that are no longer needed
fertilityrate_df = fertilityrate_df.drop(["Women 15 to 50 years!Women with births in the last 12 months", "Total Women 15 to 50 years"], axis=1)

fertilityrate_df.head()
```
